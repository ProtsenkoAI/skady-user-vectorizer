```python
import json
```

```python
with open("./data/parsed_data1.json") as f:
    cnt = json.load(f)
```

```python
with open("./data/parsed_data.jsonl", "w") as f:
    for user, user_data in cnt.items():
        f.write(json.dumps({"user_id": user, "user_data": user_data}) + "\n")
```
